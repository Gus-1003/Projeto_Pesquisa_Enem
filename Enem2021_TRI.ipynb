# Separando em grupos:

## Pela ordem das notas:

[1, 2, 3, 4, 6, 8, 9, 12, 18, 24, 36, 72, 26107, 52214, 78321, 104428, 156642, 208856, 234963, 313284, 469926, 626568, 939852 e 1879704]

```python
divisor = 3

# Calcule o tamanho do grupo baseado no número de amostras e no número de grupos desejados
tamanho_grupo = len(dados) // divisor

# Divida o DataFrame em 7 grupos com tamanhos aproximadamente iguais
grupos = [dados.iloc[i*tamanho_grupo:(i+1)*tamanho_grupo] for i in range(divisor)]

# Imprima os grupos criados
for i, grupo in enumerate(grupos):

  '''o método reset_index() do objeto DataFrame do pandas para redefinir o índice das linhas do DataFrame dados.'''
  grupo.reset_index(drop=True, inplace=True)
  
  print('GRUPO ' + str(i + 1))
  display(grupo)
  print('============================================================================')
  print('============================================================================')
```

```python
# Calcula as médias das colunas para cada grupo
medias_grupos = [grupo.mean() for grupo in grupos]

# Cria um novo dataframe com as médias das colunas para cada grupo
df_medias_grupos = pd.DataFrame(medias_grupos)
df_medias_grupos
```

```python
dados.mean()
```

```python
count_decrescente = 0

# Loop sobre cada coluna do df_medias_grupos
for coluna in df_medias_grupos.columns:
  # Cria uma figura para o gráfico de pontos
  plt.figure(figsize=(10, 5))
  
  # Plota o gráfico de pontos
  plt.scatter(x=df_medias_grupos.index, y=df_medias_grupos[coluna])
  
  # Conecta os pontos por uma linha
  plt.plot(df_medias_grupos.index, df_medias_grupos[coluna])
  
  # Define os rótulos dos eixos x e y
  plt.xlabel('Grupo')
  plt.ylabel(coluna)
  
  # Adiciona um grid de fundo
  plt.grid(True)
  
  # Exibe o gráfico na tela
  plt.show()

  is_decreasing = True
  for i in range(1, len(df_medias_grupos[coluna])):
    if df_medias_grupos[coluna][i] > df_medias_grupos[coluna][i-1]:
      is_decreasing = False
      break
          
  if not is_decreasing:
    count_decrescente =  count_decrescente + 1
```

```python
print('Número de gráficos com comportamento não decrescente:', count_decrescente)
```

```python
# Especificar os intervalos de notas desejados
intervalos = np.arange(339.7, 867.1, 30.0)

# Criar uma coluna no DataFrame para armazenar o intervalo de notas correspondente a cada amostra
dados['intervalo_notas'] = pd.cut(dados_com_notas['notas'], intervalos, right=False)

# Dividir o DataFrame em grupos baseados nos intervalos de notas
grupos = [grupo for _, grupo in dados.groupby('intervalo_notas')]

# Imprimir os grupos criados
for i, grupo in enumerate(grupos):
  print('Intervalo de notas ' +  str(i) + ': ' + str(intervalos[i]) + ' - ' + str(intervalos[i+1]))

  '''o método reset_index() do objeto DataFrame do pandas para redefinir o índice das linhas do DataFrame dados.'''
  grupo.reset_index(drop=True, inplace=True)

  display(grupo)
  print('============================================================================')
  print('============================================================================')
```

```python
# Seleciona aleatoriamente 400 amostras de cada grupo
amostras_por_grupo = 400
grupos_amostrados = []

for i, grupo in enumerate(grupos):
    if len(grupo) >= amostras_por_grupo:
        grupo_amostrado = grupo.sample(n=amostras_por_grupo, replace=False)
    else:
        grupo_amostrado = grupo.sample(n=amostras_por_grupo, replace=True)
    grupos_amostrados.append(grupo_amostrado)
```
