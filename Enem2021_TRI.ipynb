```python
# O comando pd.read_csv() é usado para ler o arquivo CSV e armazenar os dados em um objeto Pandas DataFrame chamado dados.
dados = pd.read_csv('/content/drive/MyDrive/Pesquisa_Enem2021/Base de Dados/respostas_corrigidas.csv')
```

```python
'''O comando display(dados) exibe todo o conteúdo do DataFrame na saída do notebook, permitindo visualizar todas as variáveis e observações do conjunto de dados. 
Essa função é especialmente útil quando se trabalha com DataFrames grandes ou com muitas colunas, pois evita que o notebook exiba apenas um resumo limitado dos dados.'''

display(dados)
```

```python
'''O método info() fornece informações sobre o número total de observações, o número de colunas, 
o nome das colunas, o número de valores não nulos em cada coluna e o tipo de dados de cada coluna.'''

dados.info(null_counts = True)
```

```python
# Ordene o DataFrame com base na coluna "notas", em ordem decrescente
dados = dados.sort_values(by='notas', ascending=False)

# Reset o índice do DataFrame
dados = dados.reset_index(drop=True)

# O DataFrame agora está ordenado pela coluna "notas" em ordem decrescente
display(dados)
```

```python
# O DataFrame agora não contém mais linhas com valores NaN
dados["notas"].value_counts().sum()
```

```python
# Apague as linhas com valores NaN
dados = dados.dropna()

'''o método isna() do objeto DataFrame para verificar se há valores ausentes (NaN) em cada uma das colunas do DataFrame dados. 
Em seguida, utiliza o método sum() para somar o número de valores ausentes em cada coluna.'''
dados.isna().sum()
```

```python
display(dados)
```

```python
'''o método drop() é útil para remover colunas de um DataFrame, facilitando a análise e visualização dos dados. 
O parâmetro axis=1 indica que a coluna deve ser removida e o parâmetro inplace=True indica que a alteração deve 
ser feita diretamente no DataFrame dados.'''

dados.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# Calcule o tamanho do grupo baseado no número de amostras e no número de grupos desejados
tamanho_grupo = len(dados) // 7

# Divida o DataFrame em 7 grupos com tamanhos aproximadamente iguais
grupos = [dados.iloc[i*tamanho_grupo:(i+1)*tamanho_grupo] for i in range(7)]

# Imprima os grupos criados
for i, grupo in enumerate(grupos):
  '''o método reset_index() do objeto DataFrame do pandas para redefinir o índice das linhas do DataFrame dados.'''
  dados.reset_index(drop=True, inplace=True)
     
  print(f'Grupo {i + 1}: {display(grupo)}') # Mostra apenas as primeiras linhas do grupo para fins de exemplo
```

```python
# Defina os intervalos das classes com base nas notas mínima e máxima
intervalos = [300, 401, 501, 601, 701, 801, 868]  # Adicione um valor a mais para representar a maior nota + 1

# Defina os rótulos das classes
rotulos = ['1: 300-400', '2: 401-500', '3: 501-600', '4: 601-700', '5: 701-800', '6: 801-867']

# Crie uma nova coluna no DataFrame para representar as classes
dados['classes'] = pd.cut(dados['notas'], bins=intervalos, labels=rotulos)

# Agrupe os dados por classe e calcule a frequência de cada classe
contagem_por_classe = dados['classes'].value_counts()

# Crie um gráfico de barras para mostrar a frequência de cada classe
contagem_por_classe.plot(kind='bar')
plt.title('Frequência das Classes')
plt.xlabel('Classes')
plt.ylabel('Frequência')
plt.show()
```

```python
# Calcule a frequência de cada classe em cada grupo
for i, grupo in enumerate(grupos):
      # Agrupe os dados por classe e calcule a frequência de cada classe
    contagem_por_classe = grupo['notas'].apply(lambda x: pd.cut([x], bins=intervalos, labels=rotulos)).value_counts()

    print(f'Grupo {i + 1}:')
    # Crie um gráfico de barras para mostrar a frequência de cada classe
    contagem_por_classe.plot(kind='bar')
    plt.title('Frequência das Classes')
    plt.xlabel('Classes')
    plt.ylabel('Frequência')
    plt.show()
    print('---')
```
