# Processamento 2:

## Concatenando respostas|:

```python
provaEnem = pd.concat([enemAmarela, enemAzul, enemCinza, enemRosa], ignore_index=True)
display(provaEnem)
```

## Dados das respostas:

```python
# lista com os nomes das colunas que vão ser plotadas
colunas = ['q' + str(i) for i in range(1, 46)]

# criar um DataFrame vazio para armazenar os resultados
resultado = pd.DataFrame()

# loop para iterar sobre as colunas e criar os gráficos
for coluna in colunas:
    # obter os valores de value_counts()
    valores = provaEnem[coluna].value_counts()
    porcentagens = provaEnem[coluna].value_counts(normalize=True)

    # adicionar os valores e porcentagens como colunas no DataFrame resultado
    #resultado[coluna + '_valores'] = valores
    resultado[coluna + '_porcentagens'] = porcentagens

# Ordenar as linhas em ordem alfabética pelas letras que representam as alternativas
resultado = resultado.sort_index()

# exibir o resultado final
resultado
```

```python
# Gabarito
gabarito_2022 = ['E', 'A', 'C', 'D', 'D', 'E', 'B', 'E', 'A', 'C', 'C', 'C', 'D', 'A', 'B', 'C', 'E', 'E', 'C', 'B', 'B', 'C', 'C', 'A', 'E', 'B', 'E', 'D', 'A', 'B', 'E', 'A', 'D', 'B', 'A', 'D', 'D', 'C', 'B', 'D', 'D', 'A', 'A', 'E', 'B']
# Criando um novo DataFrame com as porcentagens de acerto da questão baseado no gabarito
porcentagens_acerto = pd.DataFrame()

for questao, resposta in enumerate(gabarito, start=1):
    if resposta in resultado.index:
        porcentagens_acerto[f'q{questao}_acerto'] = [resultado.loc[resposta, f'q{questao}_porcentagens']]
    else:
        porcentagens_acerto[f'q{questao}_acerto'] = [0.0]  # Alternativa não encontrada, atribui 0.0

# Exibindo o resultado
porcentagens_acerto
```

```python
# Encontrar a coluna com o maior valor
coluna_maior_valor = porcentagens_acerto.idxmax(axis=1).iloc[0]

# Obter os rótulos das 5 colunas com os maiores valores
maiores_colunas = porcentagens_acerto.iloc[0].nlargest(6).index

# Selecionar as 5 colunas com os maiores valores
dataset_maiores_colunas = porcentagens_acerto[maiores_colunas]

print("Dataset com as 5 colunas que possuem os maiores valores:")
dataset_maiores_colunas
```

```python
# Encontrar a coluna com o maior valor
coluna_maior_valor = porcentagens_acerto.idxmax(axis=1).iloc[0]

# Obter os rótulos das 5 colunas com os maiores valores
menores_colunas = porcentagens_acerto.iloc[0].nsmallest(6).index

# Selecionar as 5 colunas com os maiores valores
dataset_menores_colunas = porcentagens_acerto[menores_colunas]

print("Dataset com as 5 colunas que possuem os menores_colunas:")
dataset_menores_colunas
```
